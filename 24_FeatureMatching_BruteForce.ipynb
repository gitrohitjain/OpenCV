```python
import cv2
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def display(i):
    plt.figure(figsize=(10,10))
    plt.imshow(i,cmap='gray')    
```

```python
book = cv2.imread('book.jpeg',0)
display(book)

slant_book = cv2.imread('slant.jpeg',0)
display(slant_book)

b=book.copy()
sb = slant_book.copy()
```

```python
orb = cv2.ORB_create()
```

```python
kp1, des1 = orb.detectAndCompute(b,None)
```

```python
kp2, des2 = orb.detectAndCompute(sb,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
```

```python
matches = bf.match(des1,des2)
```

```python
#sort on order of distance
len(matches)
```

```python
matches[0].distance,matches[1].distance 
# Less distance better match
```

```python
matches = sorted(matches, key=lambda x:x.distance)
```

```python
book_matches = cv2.drawMatches(b,kp1,sb,kp2,matches[:25],None,flags=2)
```

```python
display(book_matches)
```

```python
taj_front = cv2.imread('taj_front.jpg',0)
display(taj_front)
taj_side = cv2.imread('taj_side.jpg',0)
display(taj_side)

tf= taj_front.copy()
ts = taj_side.copy()
```

```python
orb = cv2.ORB_create()
kp1, des1 = orb.detectAndCompute(tf,None)
kp2, des2 = orb.detectAndCompute(ts,None)
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
matches = bf.match(des1,des2)
len(matches)
matches[0].distance,matches[1].distance
matches = sorted(matches, key=lambda x:x.distance)
taj_matches = cv2.drawMatches(tf,kp1,ts,kp2,matches[:25],None,flags=2)
display(taj_matches)
```
