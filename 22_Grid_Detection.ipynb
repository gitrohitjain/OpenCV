```python
import cv2
import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread('flat_chess.jpg')
plt.imshow(flat_chess)
```

```python
found, corners =cv2.findChessboardCorners(flat_chess.copy(),(8,8))

print(found, corners)
```

```python
found, corners =cv2.findChessboardCorners(flat_chess.copy(),(7,7))

print(found, corners)
```

```python
draw =cv2.drawChessboardCorners(flat_chess.copy(),(7,7), corners, found)
```

```python
plt.figure(figsize=(7,7))
plt.imshow(draw)
```

```python
dots = cv2.imread('dots.jpg')
plt.imshow(dots)
```

```python
founds, corners = cv2.findCirclesGrid(dots, 
                                      (7,7),
                                      cv2.CALIB_CB_SYMMETRIC_GRID)

founds
```

```python
draw =cv2.drawChessboardCorners(dots.copy(),(7,7), corners, found)
```

```python
plt.imshow(draw)
```
