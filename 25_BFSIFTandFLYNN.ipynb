```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def display(i):
    plt.figure(figsize=(10,10))
    plt.imshow(i,cmap='gray')    
```

```python
book_front = cv2.imread('book.jpeg',cv2.COLOR_BGR2GRAY)
display(book_front)
book_side = cv2.imread('slant.jpeg',cv2.COLOR_BGR2GRAY)
display(book_side)

bf=book_front.copy()
bs=book_side.copy()
```

```python
orb = cv2.ORB_create()
kp1, des1 = orb.detectAndCompute(bf,None)
kp2, des2 = orb.detectAndCompute(bs,None)
```

```python
bruteForce = cv2.BFMatcher()
```

```python
matches = bruteForce.knnMatch(des1,des2,k=2) # k best matches
len(matches)
```

```python
# Match of every key point
```

```python
good =[]
#Ratio Test, Less distance better Match
for match1, match2 in matches:
    if match1.distance < 0.75*match2.distance:
        good.append([match1])
```

```python
print(len(good))
good
```

```python
sift_matches = cv2.drawMatchesKnn(bf.copy(),kp1,bs,kp2,good,None,flags=2)
display(sift_matches)
```
