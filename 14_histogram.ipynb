```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def cc(i):
    return cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
```

```python
rain = cv2.imread('rainbow.jpg')
plt.imshow(cc(rain.copy()))
```

```python
blue = cv2.imread('blue.jpg')
plt.figure(figsize=(7,7))
plt.imshow(cc(blue.copy()))
```

```python
#opencv bgr
#blue is channel 0
hist_val = cv2.calcHist([blue],channels=[0],mask= None,
             histSize=[256], ranges=[0,256])
```

```python
hist_val.shape
```

```python
plt.plot(hist_val)
```

```python
#opencv bgr
#red is channel 2
hist_val = cv2.calcHist([blue],channels=[2],mask= None,
             histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_val)
```

```python
hist_val = cv2.calcHist([rain],channels=[0],mask= None,
             histSize=[256], ranges=[0,256])
plt.plot(hist_val)
```

```python
colors = ('b','g','r')

for i, col in enumerate(colors):
    histr = cv2.calcHist(rain.copy(), [i], None, [256], [0,256])
    plt.plot(histr,color=col)
```

```python
messi = cv2.imread('messi.jpg')
plt.figure(figsize=(7,7))
plt.imshow(cc(messi.copy()))
```

```python
colors = ('b','g','r')

for i, col in enumerate(colors):
    histr = cv2.calcHist(messi.copy()[0:400,400:500], [i], None, [256], [0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
```
