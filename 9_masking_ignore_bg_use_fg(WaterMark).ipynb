```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cc(i):
    return cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
```

```python
messi = cc(cv2.imread('messi.jpg'))
wm = cc(cv2.imread('wm.jpg'))
wmg = cc(cv2.imread('wm.jpg',cv2.IMREAD_GRAYSCALE))
print(messi.shape)
print(wm.shape)
```

```python
plt.imshow(wm)
```

```python
plt.imshow(wmg)
```

```python
print('b',wmg.shape)
mask = cv2.bitwise_not(wmg.copy())
print('a',mask.shape)
```

```python
plt.imshow(mask)
```

```python
fg = cv2.bitwise_or(wm.copy(),wm.copy(),mask)
plt.imshow(fg)
```

```python
fg = fg[100:300,50:250]
```

```python
plt.imshow(fg)
```
