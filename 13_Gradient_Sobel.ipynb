```python
'''
Sobel X -> removes X
Sobel Y -> removes Y
Normalized -> Keeps both
'''
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def cc(i):
    return cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
```

```python
s = cv2.imread('suduko.jpg')
plt.figure(figsize=(10,10))
plt.imshow(cc(s))
```

```python
sobelx = cv2.Sobel(s.copy(),cv2.CV_64F,1,0,ksize=31)
plt.figure(figsize=(10,10))
plt.imshow(sobelx)
```

```python
sobely = cv2.Sobel(s.copy(),cv2.CV_64F,0,1,ksize=31)
plt.figure(figsize=(10,10))
plt.imshow(sobely)
```

```python
laplacian = cv2.Laplacian(s.copy(),cv2.CV_64F,ksize=31)
plt.figure(figsize=(10,10))
plt.imshow(laplacian)
```

```python
blended = cv2.addWeighted(sobelx,0.5,sobely,0.5,0)
plt.figure(figsize=(10,10))
plt.imshow(blended)
```

```python
_,thresh = cv2.threshold(s.copy(),100,255,cv2.THRESH_BINARY)
plt.figure(figsize=(10,10))
plt.imshow(thresh)
```
