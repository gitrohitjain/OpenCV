```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cc(i):
    return cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
```

```python
#HARRIS DETECTOR

suduko = cv2.imread('suduko.jpg')[:,970:3000]
suduko_gray = cv2.imread('suduko.jpg',0)[:,970:3000]
print(suduko.shape)
plt.figure(figsize=(7,7))
plt.imshow(cc(suduko))

chess = cv2.imread('chess.jpg')
chess_gray = cv2.imread('chess.jpg', 0)
plt.figure(figsize=(6,6))
plt.imshow(cc(chess))

plt.show()
```

```python
suduko_gray
```

```python
suduko_gray = np.float32(suduko_gray)
```

```python
suduko_gray
```

```python
dst = cv2.cornerHarris(src = suduko_gray, 
                       blockSize = 2,
                       ksize = 3,
                       k = 0.04)


plt.imshow(dst)
```

```python
dst = cv2.dilate(dst,None)
plt.imshow(dst)
```

```python
dst.max()
```

```python
suduko[dst>0.01*dst.max()] = [255,0,0]
```

```python
plt.figure(figsize=(7,7))
plt.imshow(suduko)
```

```python
import numpy as np
chess_gray = np.float32(chess_gray)
chess_gray
```

```python
dst = cv2.cornerHarris(src = chess_gray, 
                       blockSize = 2,
                       ksize = 3,
                       k = 0.04)


plt.imshow(dst)
```

```python
chess[dst>0.01*dst.max()] = [255,0,0] # Mark corner
```
