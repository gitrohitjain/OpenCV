```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cc(i):
    return cv2.cvtColor(i,cv2.COLOR_BGR2RGB)

dog = cv2.imread('dog.jpg')
wm = cv2.imread('wm.jpg')
plt.imshow(cc(dog.copy()))
# plt.imshow(cc(wm.copy()))
print(dog.shape)
print(wm.shape)
```

```python
dogo = cv2.resize(dog.copy(),(180,220))
wmo = cv2.resize(wm.copy(),(180,220))
plt.imshow(cc(dogo.copy()))
print(dogo.shape)
print(wmo.shape)
```

```python
blended = cv2.addWeighted(dogo.copy(),0.5,
                          wmo.copy(),0.5,
                         gamma=0)
plt.imshow(cc(blended))
```
