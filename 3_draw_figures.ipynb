```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank = np.zeros((512,512,3), dtype = np.int16)
```

```python
blank.shape
```

```python
plt.imshow(blank)
```

```python
r = cv2.rectangle(blank.copy(), (400,10), (480,100),color=(255,0,0), thickness=5)
plt.imshow(r)
```

```python
c = cv2.circle(blank.copy(), center=(256,256), radius =50,color=(0,255,0), thickness=-1)
plt.imshow(c)
```

```python
line = cv2.line(c, (10,10), (502,502), (255,255,255), 5)
plt.imshow(line)
```
