```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cc(i):
    return cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
```

```python
suduko = cv2.imread('suduko.jpg')[:,970:3000]
suduko_gray = cv2.imread('suduko.jpg',0)[:,970:3000]
print(suduko.shape)
plt.figure(figsize=(7,7))
plt.imshow(cc(suduko))

chess = cv2.imread('chess.jpg')
chess_gray = cv2.imread('chess.jpg', 0)
plt.figure(figsize=(6,6))
plt.imshow(cc(chess))

plt.show()
```

```python
corners = cv2.goodFeaturesToTrack(suduko_gray,
                                 maxCorners = 5, #5 best corner
                                 qualityLevel =0.01, 
                                 minDistance=10)
```

```python
corners = np.int0(corners)
print(corners)
print('ravel',corners[0].ravel)
```

```python
for i in corners:
    x,y = i.ravel()
    five =cv2.circle(suduko,(x,y),20,255,-1)
```

```python
plt.imshow(five)
```

```python
corners = cv2.goodFeaturesToTrack(suduko_gray,
                                 maxCorners = 50, #50 best corner
                                 qualityLevel =0.01, 
                                 minDistance=10)

corners = np.int0(corners)
for i in corners:
    x,y = i.ravel()
    fifty =cv2.circle(suduko,(x,y),20,255,-1)

plt.imshow(fifty)
```

```python
chess = cv2.imread('chess.jpg')
chess_gray = cv2.imread('chess.jpg', 0)

cor = cv2.goodFeaturesToTrack(chess_gray,
                            maxCorners = 5, #50 best corner
                            qualityLevel =0.01, 
                            minDistance=10)

cor = np.int0(cor)

for i in cor:
    x,y = i.ravel()
    five = cv2.circle(chess,(x,y),2,255,-1)

plt.imshow(cc(five))
```
