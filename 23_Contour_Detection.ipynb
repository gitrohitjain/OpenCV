```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('contour.jpg',0)
print(img.shape)
plt.imshow(img)
```

```python
ret, thresh = cv2.threshold(img.copy(), 127,255, cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh, cmap='gray')
```

```python
ret,thresh = cv2.threshold(img.copy(),127,255,cv2.THRESH_BINARY)
```

```python
contours, hierarchy = cv2.findContours(thresh.copy(), 
                                 mode = cv2.RETR_CCOMP,
                                 method = cv2.CHAIN_APPROX_SIMPLE)
```

```python
len(contours)
```

```python
print(hierarchy)
```

```python
import numpy as np
ext_contours = np.zeros(img.shape)
```

```python
ext_contours.shape
```

```python
count = 0
for idx,item in enumerate(hierarchy[0]):
    if item[3] == -1 :
        count = count + 1
        t=cv2.drawContours(ext_contours,contours,idx,255,-1)

print('external count', count)
plt.imshow(t,cmap='gray')
```

```python
int_contours = np.zeros(img.shape)
count = 0

for idx,item in enumerate(hierarchy[0]):
    if item[3] != -1 :
        count = count + 1
        t=cv2.drawContours(int_contours,contours,idx,255,-1)

print('internal count', count)
plt.imshow(t,cmap='gray')
```
