```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cc(i):
    return cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
```

```python
taj = cv2.imread('taj.jpg').astype(np.float32)/255
plt.figure(figsize=(7,7))
plt.imshow(cc(taj))
```

```python
gamma=1/4

result = np.power(taj.copy(),gamma)
plt.imshow(cc(result))
```

```python
gamma=4

result = np.power(taj.copy(),gamma)
plt.imshow(cc(result))
```

```python
txt = cv2.putText(taj.copy(),'TAJ',(10,200),
                  fontFace= cv2.FONT_HERSHEY_COMPLEX,
                  fontScale=3, color=(0,0,255))
plt.figure(figsize=(7,7))
plt.imshow(cc(txt))
```

```python
kernel = np.ones((5,5), dtype=np.float32)/25
kernel
```

```python
f1 = cv2.filter2D(txt.copy(), -1, kernel)
plt.figure(figsize=(7,7))
plt.imshow(cc(f1))
```

```python
blurred = cv2.blur(txt.copy(), ksize=(5,5))
plt.figure(figsize=(7,7))
plt.imshow(cc(blurred))
```

```python
gauss = cv2.GaussianBlur(txt.copy(), (5,5), 10)
plt.figure(figsize=(7,7))
plt.imshow(cc(gauss))
```

```python
median = cv2.medianBlur(txt.copy(),3)
plt.figure(figsize=(7,7))
plt.imshow(cc(median))
```

```python
lady = cv2.imread('lady.jpg')
plt.figure(figsize=(7,7))
plt.imshow(cc(lady))
```

```python
median = cv2.medianBlur(lady.copy(),5)
plt.figure(figsize=(7,7))
plt.imshow(cc(median))
```

```python
bi_blur = cv2.bilateralFilter(lady.copy(), 9, 75, 75)
plt.figure(figsize=(7,7))
plt.imshow(cc(bi_blur))
```

```python
bi_blur = cv2.bilateralFilter(txt.copy(), 9, 75, 75)
plt.figure(figsize=(7,7))
plt.imshow(cc(bi_blur))
```
